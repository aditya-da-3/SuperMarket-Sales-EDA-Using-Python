```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file 'supermarket_sales.csv' and load it into a DataFrame called df_sales
df_sales = pd.read_csv("supermarket_sales.csv")

# Display the contents of the df_sales DataFrame
print(df_sales)
```

```python
# Convert the 'Date' column to datetime format
df_sales["Date"] = pd.to_datetime(df_sales["Date"])  

# Convert 'Time' to hours (extract hour from time)
df_sales['Time'] = pd.to_datetime(df_sales['Time'], format='%H:%M').dt.hour  

# Calculate the week number based on the 'Date' column
df_sales['Week_Number'] = ((df_sales['Date'] - df_sales['Date'].min()).dt.days // 7) + 1  

# Create a 'Week_Tag' column with "Week" label and corresponding week number
df_sales['Week_Tag'] = "Week " + df_sales['Week_Number'].astype(str)  

# Create a new 'Hour' column with values from the 'Time' column
df_sales['Hour'] = df_sales['Time']  
```

```python
# Group data by 'Week_Tag' and sum 'Total_Sales' for each week
df_weekly_sales = df_sales.groupby('Week_Tag')['Total_Sales'].sum().reset_index()  

# Extract week number from 'Week_Tag' and convert to integer
df_weekly_sales['Week_Number'] = df_weekly_sales['Week_Tag'].str.extract('(\d+)').astype(int)  

# Sort the data by 'Week_Number' to ensure proper chronological order
df_weekly_sales = df_weekly_sales.sort_values('Week_Number')  

# Set the figure size for the plot
plt.figure(figsize=(12, 7))  

# Create a filled area plot for sales trend
plt.fill_between(df_weekly_sales['Week_Tag'], df_weekly_sales['Total_Sales'], color='orange', alpha=0.4)  

# Plot the sales trend with line and markers
plt.plot(df_weekly_sales['Week_Tag'], df_weekly_sales['Total_Sales'], marker='o', linestyle='-', color='red')  

# Add title 
plt.title("Weekly Sales Over Time", fontsize=14, fontweight='bold')  

# Add x-axis label 
plt.xlabel("Week No", fontsize=12) 

# Add y-axis label 
plt.ylabel("Total Sales", fontsize=12)  

# Add gridlines on the y-axis 
plt.grid(axis='y', linestyle='--', alpha=0.3)  

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)  

# Set y-axis limits slightly outside the min and max values
plt.ylim(df_weekly_sales['Total_Sales'].min() - 1000, df_weekly_sales['Total_Sales'].max() + 1000)  

# Set x-axis limits to match the first and last week
plt.xlim(df_weekly_sales['Week_Tag'].iloc[0], df_weekly_sales['Week_Tag'].iloc[-1])  

# Display the plot
plt.show()  
```

```python
# Set the figure size for the plot
plt.figure(figsize=(12, 7))  

# Create a bar plot with 'Branch' on x-axis, 'Total_Sales' on y-axis, and 'Product_line' as hue (different colors for product lines)
# 'estimator=sum' aggregates the sales by summing them for each branch-product combination
# 'palette="" defines the color palette for the bars, and 'errorbar=None' removes error bars
sns.barplot(x="Branch", y="Total_Sales", hue="Product_line", data=df_sales, estimator=sum, palette="Spectral", errorbar=None)  

# Add x-axis label
plt.xlabel("Branch")  

# Add y-axis label
plt.ylabel("Total Sales")  

# Add title 
plt.title("Branch-wise Sales Performance", fontsize=14, fontweight='bold')  

# Add legend at the top-right 
plt.legend(title="Product Line", loc="upper right", prop={'size': 6})   

# Display the plot
plt.show() 
```

```python
# Group data by 'Gender' and sum 'Total_Sales' for each gender
gender_sales = df_sales.groupby("Gender")["Total_Sales"].sum()  

# Set the figure size for the pie chart
plt.figure(figsize=(7, 7))  

# Create a pie chart with 'Gender' distribution of total sales
# 'autopct='%1.1f%%'' displays the percentage on each slice
# 'startangle=90' rotates the chart
# Colors are set for each gender and wedges are outlined with specified color
plt.pie(gender_sales, labels=gender_sales.index, autopct='%1.1f%%', startangle=90, 
        colors=["lightpink", "dodgerblue"], wedgeprops={'edgecolor': 'black'})  

# Add title 
plt.title("Gender-wise Total Sales Distribution", fontsize=14, fontweight='bold')  

# Display the pie chart
plt.show()  
```

```python
# Set the figure size for the bar plot
plt.figure(figsize=(12, 7))  

# Create a bar plot with 'Payment' on x-axis, 'Total_Sales' on y-axis, and 'Customer_type' as hue (different colors for customer types)
# 'estimator=sum' aggregates the sales by summing them for each payment method-customer type combination
# 'palette=""' sets the color palette , and 'errorbar=None' removes error bars
sns.barplot(x="Payment", y="Total_Sales", hue="Customer_type", data=df_sales, estimator=sum, palette="Greys", errorbar=None)  

# Add x-axis label
plt.xlabel("Payment Method")  

# Add y-axis label
plt.ylabel("Total Sales")  

# Add title 
plt.title("Sales Performance by Customer Type & Payment Method", fontsize=14, fontweight='bold')  

# Add legend at the top-right 
plt.legend(title="Customer Type", loc="upper right", prop={'size': 10}) 

# Display the plot
plt.show()  
```

```python
# Group data by 'City' and sum 'Total_Sales' for each city
city_sales = df_sales.groupby("City")["Total_Sales"].sum()  

# Set the figure size for the pie chart
plt.figure(figsize=(7, 7))  

# Create a pie chart with 'City' distribution of total sales
# 'autopct='%1.1f%%'' displays the percentage on each slice
# 'startangle=90' rotates the chart 
# Colors are set for each city and wedges are outlined with specified color
plt.pie(city_sales, labels=city_sales.index, autopct='%1.1f%%', startangle=90, 
        colors=["lightcoral", "lightskyblue", "lightyellow"], wedgeprops={'edgecolor': 'black'})  

# Add title 
plt.title("City-wise Total Sales Distribution", fontsize=14, fontweight='bold')  

# Display the pie chart
plt.show()  
```

```python
# Set the figure size for the heatmap
plt.figure(figsize=(10, 6))  

# Create a pivot table with 'Product_line' as rows, 'Customer_type' as columns, and 'Total_Sales' as values, aggregating by sum
pivot_table = df_sales.pivot_table(index="Product_line", columns="Customer_type", values="Total_Sales", aggfunc="sum")  

# Create a heatmap from the pivot table with annotations (annot=True) to show values in each cell, formatted as integers (fmt=".0f")
# 'cmap=""' applies a specified color gradient, and 'linewidths=0.5' adds lines between the cells
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="Purples", linewidths=0.5)  

# Add title 
plt.title("Customer Type vs. Product Line Sales", fontsize=14, fontweight='bold')  

# Add x-axis label
plt.xlabel("Customer Type")  

# Add y-axis label
plt.ylabel("Product Line")  

# Display the heatmap
plt.show()  
```

```python
# Group data by 'Hour' and sum 'Total_Sales' for each hour
sales_by_hour = df_sales.groupby('Hour')['Total_Sales'].sum()  

# Set the figure size for the plot
plt.figure(figsize=(12, 6))  

# Fill the area between the x-axis and the 'Total_Sales' values with specified color and transparency
# 'step='mid'' creates a step plot where the step occurs in the middle of the range
plt.fill_between(sales_by_hour.index, sales_by_hour.values, color="lightgreen", alpha=0.5, step='mid')  

# Create a step plot with specified color, the step occurring in the middle, and a line width of 2
plt.step(sales_by_hour.index, sales_by_hour.values, where='mid', color="black", lw=2)  

# Scatter plot to mark the data points with desired color on top of the line plot (zorder=3 ensures it's on top)
plt.scatter(sales_by_hour.index, sales_by_hour.values, color="red", zorder=3)  

# Add title 
plt.title("Sales Trend Across Different Hours", fontsize=14, fontweight='bold')  

# Add x-axis label
plt.xlabel("Hour of Day")  

# Add y-axis label
plt.ylabel("Total Sales")  

# Set x-axis ticks from 10 to 20 (representing hours of the day)
plt.xticks(range(10, 21))  

# Set x-axis limits to range from 10 to 20 hours
plt.xlim(10, 20)  

# Set y-axis limits slightly outside the min and max values
plt.ylim(sales_by_hour.min() - 1000, sales_by_hour.max() + 1000)  

# Display the plot
plt.show()  
```

```python
# Set the figure size for the histogram plot
plt.figure(figsize=(8, 5))  

# Create a histogram of the 'Rating' column with 6 bins, no KDE (kernel density estimation), and color set to 'desired color'
sns.histplot(df_sales["Rating"], bins=6, kde=False, color="peachpuff")  

# Add x-axis label
plt.xlabel("Customer Rating")  

# Add y-axis label
plt.ylabel("Count")  

# Add title 
plt.title("Customer Rating Distribution", fontsize=14, fontweight='bold')  

# Display the plot
plt.show()  
```
